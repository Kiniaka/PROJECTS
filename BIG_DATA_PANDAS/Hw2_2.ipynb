```python
import pandas as pd

python_data = pd.read_csv("/content/drive/MyDrive/2017_june_final.csv")

# changing columns names from Ukrainian to English.
python_data.rename(columns={python_data.columns[7]: 'Salary change for 12 months'}, inplace=True)
python_data.rename(columns={python_data.columns[-2]: 'Currency'}, inplace=True)

#created new DataFrame, which only contains lines with specialists who have given the Python programming language
python_data = python_data.dropna(subset=['Programming language'])

# presented all columns and number of empty rows in each column (0 - means that there is no empty row)
python_data.isnull().sum()
```

```python
# Quantity of all rows in python_data

print(python_data.size)
```

```python
# Grouping by 'Specialisation'column and don't show empty rows.

print(python_data.groupby(by = ['Specialisation '], dropna = True))
```

```python
# Present list of max and min salary by month per Role

salary = python_data.groupby('Role').agg({'Salary/month': ['min', 'max']})

print(salary)
```

```python
# Calculated average salary per month for each type of role:

g1 = python_data.groupby('Role', group_keys=True)
fill_avg_salary_per_role = g1['Salary/month'].apply(lambda x: x.mean().round(2))

# Data to check if the salary average has been calculated correctly:

# suma = g1['Salary/month'].apply(lambda x: x.sum())
# srednia = g1['Salary/month'].apply(lambda x: ))
# print(suma)
# print(srednia)

print('Average salary in a each role')
print('______________________________')
print(fill_avg_salary_per_role)

# Added new column called "avg":
python_data["avg"] = 'NaN'
```

```python
# Using describe() function to see parametres like(count - quantity rows per columns, mean -  averager per columns)
python_data.describe()
```

```python
# Saving data in new csv file  named "new_python_file.csv":

python_data.to_csv("/content/drive/MyDrive/new_python_file.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Role': [
        'Data Scientist / Data Analyst',
        'DevOps',
        'ERP / CRM',
        'HTML coder',
        'Junior Software Engineer',
        'Product Manager',
        'Senior Software Engineer',
        'Software Engineer',
        'System Architect',
        'Team lead',
        'Technical Lead'
    ],
    'min': [2700, 7000, 2000, 385, 75, 3000, 350, 100, 400, 2500, 500],
    'max': [2700, 7000, 2000, 600, 2600, 3000, 9200, 7000, 10000, 4000, 18000]
}

df = pd.DataFrame(data)
df.set_index('Role', inplace=True)

# Tworzenie wykresu liniowego
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['min'], marker='o', label='Min')
plt.plot(df.index, df['max'], marker='x', label='Max')
plt.ylabel('Salary per month (PLN)')
plt.title('SALARY PER MONTH FOR EACH PROFFESSION')
plt.xticks(rotation=80)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
